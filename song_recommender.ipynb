```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_songs =df_songs[df_songs['track_popularity'] >=60]
```

```python
df_songs.shape
```

```python
cv = CountVectorizer(max_features=5000)
vector = cv.fit_transform(df_songs['tags']).toarray()
```

```python
list(cv.get_feature_names_out())
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(song):
    song_index = df_songs[df_songs['track_name'] == song].index[0]
    dist = similarity[song_index]
    song_list = sorted(list(enumerate(dist)),reverse=True, key=lambda x: x[1])[1:10]
    for i in song_list:
        print(df_songs.iloc[i[0]].track_name)
```

```python
recommend('Memories - Dillon Francis Remix')
```
